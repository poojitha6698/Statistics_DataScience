### Nominal Data

These represent labels without any ranking.

show_id

type (Movie / TV Show)

title

director

cast

country

listed_in (genres)

description

#### Why important?

Nominal features help us segment the content: by country, director, or genre.

### Ordinal Data

Only one column fits here:

rating (G, PG, PG-13, TV-MA etc.)

Ratings have a natural order of audience maturity—making them ordinal.

### Continuous Data

(None in this dataset) – Netflix titles dataset has no numeric columns with floating or continuous value ranges.

### Discrete Data

release_year

duration (minutes for movies, number of seasons for TV shows)

date_added (can be broken into year, month, day)

These help us track how content evolved with time.
